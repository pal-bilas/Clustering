### Feature sleection for the model

Normalization is a statistical method that helps mathematical-based algorithms to interpret features with different magnitudes and distributions equally. We use StandardScaler() to normalize our dataset.

```python
from sklearn.preprocessing import StandardScaler
```

```python
X = StandardScaler().fit_transform(df1.values)
```

```python
print(X[0:5])
```

### First K Means clustering

**Let's utilize Elbow method to decide the number of cluster.**

```python
from sklearn.cluster import KMeans
distortion = []
K = range(1,11)

for k in K:
    kmeanModel=KMeans(n_clusters=k, init='k-means++', random_state=0).fit(X)
    kmeanModel.fit(X)
    distortion.append(kmeanModel.inertia_)
```

```python
plt.plot(K, distortion, 'bo-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

Elbow shows k=3 is a good choice!

```python
kmeans = KMeans(n_clusters=3, init='k-means++', n_init=100)
kmeans.fit(X)
```

```python
labels = kmeans.labels_
print(labels)
```

```python
df['clus_label'] = labels
```

```python
kmeans.cluster_centers_
```

**K Means clustering visualization**

```python
data1 = dict(type='choropleth',
            #colorscale = 'YlGnBu',
            locations = df['Country'],
            z = df['clus_label'],
            locationmode = 'country names',
            text = df['Country'],
            colorbar = {'title':"Happiness index"}
            ) 
```

```python
layout1 = dict(title = 'Happiness Index 2017 by K Means Clustering', geo = dict(showframe = False))
```

```python
choromap1 = go.Figure(data = [data1], layout=layout1)
iplot(choromap1)
```

```python
plt.scatter(x=df['Economy..GDP.per.Capita.'], y=df['Trust..Government.Corruption.'], c=kmeans.labels_)
plt.xlabel('GDP per Capita', fontsize=16)
plt.ylabel('Corruption', fontsize=16)
```

K means clustering shows a good sense of the world happiness.

**Lets now try Agglomerative Clustering**

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
agg = AgglomerativeClustering(n_clusters=3)
```

```python
agg.fit(X)
```

```python
agg.labels_
```

```python
plt.scatter(x=df['Economy..GDP.per.Capita.'], y=df['Trust..Government.Corruption.'], c=agg.labels_)
plt.xlabel('GDP per Capita', fontsize=16)
plt.ylabel('Corruption', fontsize=16)
```

```python
df['agg_labels'] = agg.labels_
```

```python
data2 = dict(type='choropleth',
            #colorscale = 'YlGnBu',
            locations = df['Country'],
            z = df['agg_labels'],
            locationmode = 'country names',
            text = df['Country'],
            colorbar = {'title':"Agglomerative index"}
            ) 
```

```python
layout2 = dict(title = 'Happiness Index 2017 by Agglomerative Clustering', geo = dict(showframe = False))
```

```python
choromap2 = go.Figure(data = [data2], layout=layout2)
iplot(choromap2)
```

Got quite a similar results, but as clustering is unsupervised ML, we can not tell you that which one is best. We will explore more in future.
